```python
import simulation as sim

import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
cauchy_data = sim.generate_cauchy_data()
sim.plot_histogram(cauchy_data)
```

```python
data = cauchy_data


fig, axes = plt.subplots(1, 2, figsize=(15, 10))

AIC_norm, BIC_norm, params = sim.fit_norm(data)
sim.plot_norm(data, **params , ax=axes[0])

AIC_t, BIC_t, params_t = sim.fit_t(data)
sim.plot_t(data, **params_t, ax=axes[1])

fig, axes = plt.subplots(1, 3, figsize=(25, 10))

AIC_gmm2, BIC_gmm2, gmm = sim.fit_gmm(data, n_components=2)
sim.plot_gmm(data, gmm, n_components=2, ax=axes[0])

AIC_gmm3, BIC_gmm3, gmm = sim.fit_gmm(data, n_components=3)
sim.plot_gmm(data, gmm, n_components=3, ax=axes[1])

AIC_gmm4, BIC_gmm4, gmm = sim.fit_gmm(data, n_components=4)
sim.plot_gmm(data, gmm, n_components=4, ax=axes[2])
```
